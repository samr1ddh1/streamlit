```python
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import concatenate
from tensorflow.keras.layers import MaxPooling1D
from tensorflow.keras.layers import Conv1D
from tensorflow.keras.layers import Dropout , LSTM
from keras.layers import  CuDNNLSTM
```

```python

data=pd.read_csv('city_hour.csv')
data.head()
```

```python
data.drop(["Xylene","NH3","Toluene","NH3","Benzene"],axis=1,inplace = True)
```

```python
data1=data.dropna()
```

```python
import missingno as msno

msno.matrix(data1)
```

```python
X_data=pd.DataFrame(data1.iloc[:,2:10].values)
Y_data=pd.DataFrame(data1.iloc[:,10].values)
```

```python
X_data
```

```python
Y_data
```

```python
sc=StandardScaler()
X_data=sc.fit_transform(X_data)
X_train,X_test,Y_train,Y_test=train_test_split(X_data,Y_data,test_size=0.2,random_state=0)
Y_train.shape
```

```python
from keras.models import Sequential

from keras.layers import Dense, Dropout

input_size = 8

model = Sequential()
model.add(LSTM(128, input_shape=(8, 1), activation='relu',return_sequences=True))
model.add(LSTM(128, activation='relu',return_sequences=True))


model.add(LSTM(64))


model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1))

model.compile(loss='mse')
```

```python
deep_model = Sequential()
input_size=8;
deep_model.add(Dense(256, input_dim=input_size, activation='relu'))

deep_model.add(Dense(256, activation='relu'))

deep_model.add(Dense(128, activation='relu'))

deep_model.add(Dense(128, activation='relu'))

deep_model.add(Dense(64, activation='relu'))

deep_model.add(Dense(64, activation='relu'))

deep_model.add(Dense(32, activation='relu'))

deep_model.add(Dense(8, activation='relu'))

deep_model.add(Dense(1))

deep_model.compile(loss='mse')
```

```python
from keras.models import Sequential

from keras.layers import Dense, Dropout

input_size = 8

cnn_model = Sequential()
cnn_model.add(Conv1D(64, kernel_size=2, activation='relu',
                     input_shape=(8, 1), padding='same'))

cnn_model.add(MaxPooling1D(2, padding='same'))
cnn_model.add(Conv1D(32, 3, activation='linear', padding='same'))
cnn_model.add(MaxPooling1D(pool_size=2, padding='same'))
cnn_model.add(Conv1D(32, 3, activation='relu', padding='same'))
cnn_model.add(MaxPooling1D(pool_size=2, padding='same'))
cnn_model.add(Flatten())
cnn_model.add(Dense(50, activation='relu'))

cnn_model.add(Dense(1))

cnn_model.compile(loss='mse', metrics=['accuracy'])
```

```python
cnn_model.fit(X_train, Y_train, 
               epochs=300,batch_size=512)
```

```python
deep_model.fit(X_train, Y_train, 
               epochs=500,batch_size=512)
```
